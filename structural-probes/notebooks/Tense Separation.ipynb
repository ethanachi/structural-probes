```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
representations = np.load('/u/scr/ethanchi/embeddings/larger/representations.npy')
projection_vector = np.load('/u/scr/ethanchi/embeddings/larger/representations-vector.npy')

verbs = representations[:, 2]
```

```python
dim = representations.shape[-1]
print(projection_vector.shape, verbs.shape)
```

```python
plt.figure(figsize=(5, 5))

plt.scatter(verbs[::2] @ projection_vector, np.zeros(verbs[::2].shape[0]), label="present")
plt.scatter(verbs[1::2] @ projection_vector, np.zeros(verbs[1::2].shape[0]), label="past")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(5, 5))

plt.scatter(verbs[::2] @ projection_vector, np.zeros(verbs[::2].shape[0]), label="present", s=0.01)
plt.scatter(verbs[1::2] @ projection_vector, np.zeros(verbs[1::2].shape[0]), label="past", s=0.01)

plt.show()
```
